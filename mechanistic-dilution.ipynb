```python
# -*- coding: utf-8 -*-

from IPython.display import display,Math,Markdown
from sympy import *

t,F,rho,V = symbols('t,F,rho,V', positive=True) #Define constants
CA = symbols('CA', cls=Function) # Define variable

deqn = Eq(V*diff(CA(t),t), -F*CA(t)) # Define equation

sol = dsolve(deqn) #Solve equation
sol # report analytical solution

init_cond = {CA(0):0.0227} # initial value of concentration.

par = [(V,440000),(F,40000),(rho,1000)]
sol = dsolve(deqn.subs(par), ics=init_cond)
sol
#Plotting the solution for inspection. Does it make sense? Connect the mathematical model to reality.

plot(sol.rhs,(t,0,5))
#evaluate the concentration after 5 hours

sol.rhs.evalf(subs={t: 5})
```
